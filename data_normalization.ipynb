Data Normalization

```
import pandas as pd
import numpy as np
```

```
# Normalize prepared data
# Load data from repo
test_df = pd.read_csv('https://raw.githubusercontent.com/treyra/CS155_miniproject/data_visualization_v1/data_cleanup/testdata_imputed.csv',sep=',', header=0,index_col=0)
train_df = pd.read_csv('https://raw.githubusercontent.com/treyra/CS155_miniproject/data_visualization_v1/data_cleanup/traindata_imputed.csv',sep=',', header=0,index_col=0)
# normalize data
test_df.head()

#--- First stuff with same scales
# change state categogy 2 to -1 - states in [1,-1]
train_df['STATE']=train_df['STATE'].replace([2],-1)
test_df['STATE']=test_df['STATE'].replace([2],-1)
# lat lon in radians first
train_df['LATITUDE']=np.deg2rad(train_df['LATITUDE'])
train_df['LONGITUDE']=np.deg2rad(train_df['LONGITUDE'])
test_df['LATITUDE']=np.deg2rad(test_df['LATITUDE'])
test_df['LONGITUDE']=np.deg2rad(test_df['LONGITUDE'])
# discovery time in interval of [0,1] day
train_df['DISCOVERY_TIME']= train_df['DISCOVERY_TIME']/2400
test_df['DISCOVERY_TIME']= test_df['DISCOVERY_TIME']/2400
# day in [0,1] 
train_df['DAY']= train_df['DAY']/366
test_df['DAY']= test_df['DAY']/366
# normalize year in [0,1], where 0 is 1992 and 1 is 2010 - MinMax scaled by hand
train_df['YEAR']= (train_df['YEAR']-1992)/(2009-1992)
test_df['YEAR']= (test_df['YEAR']-1992)/(2009-1992)

#--- Now train normalizers on training dataframe
train_df_to_norm = train_df.drop(['id','STATE','DISCOVERY_TIME','DAY','YEAR','LABEL'], axis=1)
test_df_to_norm = test_df.drop(['id','STATE','DISCOVERY_TIME','DAY','YEAR'], axis=1)
#
from sklearn.preprocessing import MinMaxScaler
minmax_scaler = MinMaxScaler()
traindata_minmax = minmax_scaler.fit_transform(train_df_to_norm)
testdata_minmax = minmax_scaler.transform(test_df_to_norm)
#
from sklearn.preprocessing import RobustScaler
robust_scaler = RobustScaler()
traindata_robust = robust_scaler.fit_transform(train_df_to_norm)
testdata_robust = robust_scaler.transform(test_df_to_norm)
#
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
traindata_standard = scaler.fit_transform(train_df_to_norm)
testdata_standard = scaler.transform(test_df_to_norm)
#
# Now put back into the dataframes where it belongs
traindata_normalized_minmax = train_df.copy()
traindata_normalized_minmax['LATITUDE']=traindata_minmax[:,0]
traindata_normalized_minmax['LONGITUDE']=traindata_minmax[:,1]
traindata_normalized_minmax['FIRE_SIZE']=traindata_minmax[:,2]
traindata_normalized_minmax['FIPS_NAME']=traindata_minmax[:,3]
traindata_normalized_minmax['FIPS_CODE']=traindata_minmax[:,4]
traindata_normalized_minmax['SOURCE_REPORTING_UNIT_NAME']=traindata_minmax[:,5]
#
traindata_normalized_robust = train_df.copy()
traindata_normalized_robust['LATITUDE']=traindata_robust[:,0]
traindata_normalized_robust['LONGITUDE']=traindata_robust[:,1]
traindata_normalized_robust['FIRE_SIZE']=traindata_robust[:,2]
traindata_normalized_robust['FIPS_NAME']=traindata_robust[:,3]
traindata_normalized_robust['FIPS_CODE']=traindata_robust[:,4]
traindata_normalized_robust['SOURCE_REPORTING_UNIT_NAME']=traindata_robust[:,5]

traindata_normalized_standard = train_df.copy()
traindata_normalized_standard['LATITUDE']=traindata_standard[:,0]
traindata_normalized_standard['LONGITUDE']=traindata_standard[:,1]
traindata_normalized_standard['FIRE_SIZE']=traindata_standard[:,2]
traindata_normalized_standard['FIPS_NAME']=traindata_standard[:,3]
traindata_normalized_standard['FIPS_CODE']=traindata_standard[:,4]
traindata_normalized_standard['SOURCE_REPORTING_UNIT_NAME']=traindata_standard[:,5]
# testdata
testdata_normalized_minmax = test_df.copy()
testdata_normalized_minmax['LATITUDE']=testdata_minmax[:,0]
testdata_normalized_minmax['LONGITUDE']=testdata_minmax[:,1]
testdata_normalized_minmax['FIRE_SIZE']=testdata_minmax[:,2]
testdata_normalized_minmax['FIPS_NAME']=testdata_minmax[:,3]
testdata_normalized_minmax['FIPS_CODE']=testdata_minmax[:,4]
testdata_normalized_minmax['SOURCE_REPORTING_UNIT_NAME']=testdata_minmax[:,5]
#
testdata_normalized_robust = test_df.copy()
testdata_normalized_robust['LATITUDE']=testdata_robust[:,0]
testdata_normalized_robust['LONGITUDE']=testdata_robust[:,1]
testdata_normalized_robust['FIRE_SIZE']=testdata_robust[:,2]
testdata_normalized_robust['FIPS_NAME']=testdata_robust[:,3]
testdata_normalized_robust['FIPS_CODE']=testdata_robust[:,4]
testdata_normalized_robust['SOURCE_REPORTING_UNIT_NAME']=testdata_robust[:,5]
#
testdata_normalized_standard = test_df.copy()
testdata_normalized_standard['LATITUDE']=testdata_standard[:,0]
testdata_normalized_standard['LONGITUDE']=testdata_standard[:,1]
testdata_normalized_standard['FIRE_SIZE']=testdata_standard[:,2]
testdata_normalized_standard['FIPS_NAME']=testdata_standard[:,3]
testdata_normalized_standard['FIPS_CODE']=testdata_standard[:,4]
testdata_normalized_standard['SOURCE_REPORTING_UNIT_NAME']=testdata_standard[:,5]
# #
traindata_normalized_minmax.head()
```

```
# adding one hot encoding for training data
# now make one hot encoder from label for the training date
from sklearn.preprocessing import OneHotEncoder
# creating instance of one-hot-encoder
enc = OneHotEncoder(handle_unknown='ignore')
# passing bridge-types-cat column (label encoded values of bridge_types)
enc_df = pd.DataFrame(enc.fit_transform(train_df[['LABEL']]).toarray())
# merge with main df bridge_df on key values
traindata_normalized_minmax = traindata_normalized_minmax.join(enc_df)
traindata_normalized_robust = traindata_normalized_robust.join(enc_df)
traindata_normalized_standard = traindata_normalized_standard.join(enc_df)
traindata_normalized_minmax.head()
```
